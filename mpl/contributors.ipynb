```python
import pandas as pd
```

```python
df = pd.read_csv('author_names.txt', header=None)
```

```python
cdf = df[0].value_counts().to_frame().reset_index()
```

```python
cdf
```

```python
def get_name(name):
    if " " in name:
        parts = name.split(" ")
        name = parts[0] + " " + parts[-1][0]
    return name
```

```python
get_name("Hannah Aizenman")
```

```python
cdf['first'] = cdf['index'].apply(get_name)
```

```python
text = dict(zip(cdf['first'].values, cdf[0].values))
```

```python
from wordcloud import WordCloud,  ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
import matplotlib.colors as mcolors
```

```python
import matplotlib.cm as mcm
colors = [(*mcm.jet(r / 10.)[:3],0.6) for r in np.array([2, 6, 8, 7, 4, 5, 8])]
```

```python
cmap = mcolors.ListedColormap(['#11557c']+colors)
```

```python
wc = WordCloud(background_color='white', max_words=1000)
wc.generate_from_frequencies(text)

# show
fig, ax = plt.subplots(figsize=(20,10))
_ = ax.imshow(wc.recolor(colormap=cmap), interpolation="bilinear")
_ = ax.axis("off")
```
